Init TTS model

```python
import torch
from TTS.api import TTS
from IPython.display import Audio
```

```python
# Get device
device = "cuda" if torch.cuda.is_available() else "cpu"
```

```python
# List available TTS models
print(TTS().list_models())
```

**Running a single speaker model**


TTS in English

```python
# Init TTS with the target model name
tts = TTS(model_name="tts_models/en/ljspeech/tacotron2-DDC", progress_bar=False).to(device)
```

```python
# Run TTS
tts.tts_to_file(text="A ball is thrown straight up at a speed of 20 meters per second. As it goes up, gravity slows it down until it stops for a moment at the highest point.We can figure out how high it goes using a formula from physics. After doing the math, we find: The ball will reach a maximum height of about 20.4 meters before it starts to fall back down.", file_path="./output.wav")
```

```python
Audio("./output.wav")
```

TTS in Mandarin

```python
tts = TTS(model_name="tts_models/multilingual/multi-dataset/xtts_v2", progress_bar=False).to(device)
```

```python
tts.tts_to_file(text="مرحبًا، هذا مثال على تحويل النص إلى كلام باللغة العربية",file_path="./output1.wav")
```
